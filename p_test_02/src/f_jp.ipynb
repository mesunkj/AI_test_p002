```python



from huggingface_hub import InferenceClient
import os
import time
from concurrent.futures import ThreadPoolExecutor, as_completed

# 初始化 InferenceClient，指定 Hugging Face 提供者和 Token
client = InferenceClient(
    provider="hf-inference",
    token=token  # 用你的 Hugging Face Token 替換
)

# 圖片儲存資料夾
OUTPUT_DIR = "../output_images"
os.makedirs(OUTPUT_DIR, exist_ok=True)



# ✅ 使用 `InferenceClient` 生成單張圖片
def generate_image(prompt: str, index: int) -> str:
    print(f"[{index}] Sending request: {prompt[:40]}...")

    # 使用 `InferenceClient` 進行推理
    image = client.text_to_image(
        prompt=prompt,
        negative_prompt=NEGATIVE_PROMPT,
        width=768,
        height=512,
        num_inference_steps=30,
        guidance_scale=7.5,
        model="stabilityai/stable-diffusion-3.5-large"

    )

    # 儲存圖片
    id_ = f'_{index}'
    filename = f"../output_images/pic_{id_}.png"
    image.save(filename)
    
    print(f"[{index}] Image saved: {filename}")
    return filename

# ✅ 批次並行處理
def generate_batch(prompts: list, max_workers=3):
    with ThreadPoolExecutor(max_workers=max_workers) as executor:
        futures = []
        for idx, prompt in enumerate(prompts):
            futures.append(executor.submit(generate_image, prompt, idx))

        for future in as_completed(futures):
            _ = future.result()
    print("done")  # 所有任務完成後印出

# ✅ 主程式
# if __name__ == "__main__":
#     generate_batch(PROMPTS, max_workers=3)


generate_batch(PROMPTS, max_workers=3)
```
